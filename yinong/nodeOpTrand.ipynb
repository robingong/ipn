```python
#网点存款 平均数 曲线
dfLess1000 = df.loc[(df[u'存款数2']<1000)]
#dfLess1000.plot(x=u'签约时间2',y=u'存款数2',kind='scatter')#开业时间2\

#
dfLess1000[u'签约时间3'] = dfLess1000[u'签约时间2']
grouped = dfLess1000.groupby(u'签约时间2').mean()

print grouped
meanCash = grouped.loc[:,[u'签约时间3',u'存款数2']]
print meanCash


plt.plot(dfLess1000[u'签约时间2'],dfLess1000[u'存款数2'],'r.',
        meanCash[u'签约时间3'],meanCash[u'存款数2'],'g-')
plt.show()

dfLess1000['signTime3'] = dfLess1000[u'签约时间3']
dfLess1000['cash2'] = dfLess1000[u'存款数2']
```

```python
#网点存款 平均数 曲线 非常波动
#http://jingyan.baidu.com/article/2a138328b257f1074b134f49.html
#http://statsmodels.sourceforge.net/stable/example_formulas.html
import statsmodels.formula.api as smf

mod = smf.ols('signTime3~cash2',data=dfLess1000)
res = mod.fit()
print res.summary()
print res.params


plt.plot(dfLess1000[u'签约时间2'],dfLess1000[u'存款数2'],'r.',
        meanCash[u'签约时间3'],meanCash[u'存款数2'],'g-',
        dfLess1000['signTime3'],res.fittedvalues,'b-')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)
def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

ax.plot_surface(X,Y,f(X,Y), rstride=1, cstride=1)

plt.show()
```
