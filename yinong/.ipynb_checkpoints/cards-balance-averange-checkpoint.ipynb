```python
data = expData
#创建一个新函数:  
def num_missing(x):
    return sum(x.isnull())  
#Apply到每一列:  
print "Missing values per column:"  
print data.apply(num_missing, axis=0) #axis=0代表函数应用于每一列  
#Apply到每一行: 
print "nMissing values per row:"  
print data.apply(num_missing, axis=1).head() #axis=1代表函数应用于每一行 
```
